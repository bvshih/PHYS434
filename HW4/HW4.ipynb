```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
plt.rcParams["figure.figsize"] = (15,10)
```

```python
t = np.linspace(0,np.pi*100,int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=int(1e5))*np.sqrt(N)+N
```

# Problem 1

```python
y = bkgd[:1000]
x = t[:1000]
plt.scatter(x, y)
plt.show()
```

```python
start = 500
points = 2000

y = bkgd[start:start+points]
x = t[start:start+points]

plt.scatter(x,y)
plt.show()
```

```python
step = 100

y = bkgd[::step]
x = t[::step]

plt.scatter(x,y)
plt.show()
```

b) The colors represent the number of signals in each bin. For example in the bottom plot, the brightest yellow indicates there are 160 values that are in the bin. The larger the bin number, the smaller size of the bin. So in the top plot, there is a maximum of 55 values in a bin but in the bottom plot there is a maximum of 160 values in a bin.
